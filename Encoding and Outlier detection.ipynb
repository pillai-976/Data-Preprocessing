```python
#Importing libraries

import pandas as pd
import numpy as np
import seaborn as sn
%matplotlib inline
import matplotlib.pyplot as plt
pd.set_option("display.max_columns",None)
df = pd.read_csv('cleaned_data',index_col='Unnamed: 0')
df
```

```python
df.info()
```

```python
print("Categorical Labels -:")
categorical_feature = [feature for feature in df.columns if df[feature].dtype == 'O']
for feature in categorical_feature:
    print(feature,' : ',len(df[feature].unique()),'labels')
```

```python
print("Numeric Labels -:")
categorical_feature = [feature for feature in df.columns if df[feature].dtype != 'O']
for feature in categorical_feature:
    print(feature,' : ',len(df[feature].unique()),'labels')
```

```python
print("Categories in each feature are as follows :\n")
for feature in df.columns:
    if df[feature].dtype == 'object':
        print(feature,'--\n')
        print(df[feature].value_counts())
        print('********************************************************************************************************')
```

```python
df['Type']
```

```python
# h - house,cottage,villa, semi,terrace;
# u - unit, duplex;
# t - townhouse;

# Labelencoding is used for attribute 'Type'

from sklearn.preprocessing import LabelEncoder     #h:0,t:1,u:2

le = LabelEncoder()
df['Type'] = le.fit_transform(df['Type'])
df['Type']
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Date:year']=df['Date'].dt.year
df.drop('Date',axis=1,inplace=True)
df.head()
```

```python
# Target guided encoding

# Suburb  :  351 labels
# Address  :  34009 labels
# Type  :  3 labels
# Method  :  9 labels
# SellerG  :  388 labels
# Date  :  78 labels
# CouncilArea  :  33 labels
# Regionname  :  8 labels

features = ['Suburb','Address','Method','SellerG','CouncilArea','Regionname']
for feature in features:
    feature_price = df.groupby([feature])['Price'].median().sort_values()
    feature_encoded = feature_price.index
    Encoded_feature_labels = { k:i for i, k in enumerate(feature_encoded, 0) }
    df[feature] = df[feature].map(Encoded_feature_labels)
    
df.head()
```

```python
df.groupby(['Date:year'])['Price'].max().plot.bar()
plt.xlabel('Date:year')
plt.xticks(rotation=360)
plt.ylabel('Price')
plt.title('Year V/s Price')
plt.show()
```

```python
df['Time'] = df['Date:year'] - df['YearBuilt']
```

## Outlier Detection

```python
df.columns
```

```python
# Plotting boxplot of all features

def plot_boxplot(df,features):
     for feature in features:
            sn.boxplot(x=df[feature])
            plt.show()

plot_boxplot(df,df.columns)
```

```python
# Normal Distribution of all features

import scipy.stats as stat
import pylab
def plot_data(df,features):
    for feature in features:
        plt.figure(figsize=(10,6))
        plt.subplot(1,2,1)
        plt.xlabel(feature)
        df[feature].hist()
        plt.subplot(1,2,2)
        stat.probplot(df[feature],dist='norm',plot=pylab)
        plt.xlabel(feature)
        plt.show()
    
plot_data(df,df.columns)
```

```python
# Logarithmic transform 

def plot_data_log(df,feature):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    plt.xlabel(feature)
    df[feature].hist()
    plt.subplot(1,2,2)
    stat.probplot(df[feature],dist='norm',plot=pylab)
    plt.xlabel(feature)
    plt.show()
```

```python
df['YearBuilt_log'] = np.log(df['YearBuilt'])
plot_data_log(df,'YearBuilt_log')
```

```python
df['Propertycount_log'] = np.log(df['Propertycount'])
plot_data_log(df,'Propertycount_log')
```

## Removing Outliers

```python
sn.boxplot(x=df['Rooms'])
```

```python
Q1 = df["Rooms"].quantile(0.25)                               #Q1 is the first quartile and Q3 is the third quartile.
Q3 = df["Rooms"].quantile(0.75)
IQR = Q3 - Q1
print("IQR:",IQR)

# find the upper fence and lower fence and print all the data above the upper fence and below the lower fence. 

Lower_Fence = Q1 - (1.5 * IQR)
Upper_Fence = Q3 + (1.5 * IQR)
print("Lower_Fence:",Lower_Fence)
print("Upper_Fence:",Upper_Fence)

# Filter out the outlier data and print only the potential data. 
# Negate the preceding result using the ~ operator:

df = df[~((df ["Rooms"] < Lower_Fence) |(df["Rooms"] > Upper_Fence))]

df
```
